```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import datetime
```

```
df1 = pd.read_csv('epl_players.csv', parse_dates=['date'])
```

```
df1['age'] = pd.to_datetime(df1.birthday)
```

```
df1.head()
```

```
def from_dob_to_age(born):
    return 2014 - born.year - ((1, 1) < (born.month, born.day))
```

```
df1['age'] = df1['age'].apply(lambda x: from_dob_to_age(x))
```

```
df1.head()
```

```
print(datetime.date.today())
```

```
start_date = '2014-01-01'
end_date = '2015-12-31'
mask =  (df1['date'] <= start_date)
```

```
df = df1.loc[mask]
```

```
df.describe()
```

```
df['goal_keeping_overall'] = (df['gk_diving'] + df['gk_handling'] + df['gk_kicking'] + df['gk_positioning'] + df['gk_reflexes']) / 5 
df['tackle_skill'] = (df['sliding_tackle'] + df['standing_tackle']) / 2
df['crossing'] = (df['crossing'] + df['finishing'])/2
```

```
df.head()
```

```
df_removed = df.drop(['birthday','curve','date','acceleration','gk_diving','sprint_speed','finishing','gk_handling', 'gk_kicking' ,'gk_positioning','gk_reflexes','standing_tackle' ,'sliding_tackle' ,'birthday',
        'preferred_foot','agility','reactions','interceptions','positioning','vision'], axis=1)
```

```
df_removed['attacking_work_rate'] = df_removed['attacking_work_rate'].apply(lambda x: 3 if x =='high' else (2 if x =='medium' else 1))  
df_removed['defensive_work_rate'] = df_removed['defensive_work_rate'].apply(lambda x: 3 if x =='high' else (2 if x =='medium' else 1)) 
```

```
df_removed.head()
```

```
clean_daataset = final_df.groupby(['player_id']).mean()
```

```
clean_daataset.head()
```

```
clean_daataset
```

```
goals_data = pd.read_csv("epl_goals.csv")
```

```
goals_data.head()
```

```
goals_data['goal_number'] = (goals_data['goal_number'] > 0).astype(int)
```

```
goals_data.drop(['goal_type','match_id','team_id' ,'player2_id'],axis=1)
```

```
goal_data = goals_data.groupby(['player_id']).sum()
```

```
goal_data.head()
```

```
goal_data = goal_data.drop(['match_id','team_id' ,'player2_id'],axis=1)
```

```
goal_data.head()
```

```
my_fully_final_data = pd.merge(goal_data, clean_daataset, on='player_id')
```

```
my_fully_final_data.fillna(0)
```

```
my_fully_final_data.head()
```

```
##Metric to deside valuable players

my_fully_final_data['My_metric'] = 100 * my_fully_final_data['goal_number'] + 50 * (my_fully_final_data['attacking_work_rate'] + my_fully_final_data['defensive_work_rate'] + my_fully_final_data['shot_power']) + 30 * (my_fully_final_data['short_passing']  + my_fully_final_data['heading_accuracy']) + 20 * (my_fully_final_data['dribbling'] + my_fully_final_data['free_kick_accuracy']) + 10 * (my_fully_final_data['long_passing'] + my_fully_final_data['ball_control']) + 5 * (my_fully_final_data['jumping'] + my_fully_final_data['stamina'] + my_fully_final_data['strength']) - 50 * (my_fully_final_data['aggression']  + my_fully_final_data['penalties'] + my_fully_final_data['age']) 
```

```
my_fully_final_data.head()
```

```
result = my_fully_final_data[my_fully_final_data['My_metric'].notna()]
```

```
answer = result.sort_values('My_metric')
```

```
least_valuable = answer.head(10)
most_valuable = answer.tail(10)
```
